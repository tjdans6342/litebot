**1. Robot Library**<br>

```python
#Robot Class package
#/home/jetson/Setup/venv/lib/python3.8/site-packages/tiki/mini/__init__.py

from tiki.mini import TikiMini
tiki = TikiMini()
import time
```

**2. Battery Check**<br>

```python
#Battery Check, return voltage(V)
tiki.get_battery_voltage()
```

```python
#Battery Checck, return current(A)
tiki.get_current()
```

**3. OLED Display**<br>

```python
#OLED Display
tiki.log("HELLO!") 
```

```python
#OLED Log Clear
tiki.log_clear()
```

**4. LED Strip**<br>

```python
#set_led(direction, index, R, G, B) 
#direction : 0(top 16-bit LED Strip), 1(left 8-bit LED), 2(right 8-bit LED)
#index : LED number(0~15 for top, 0~7 for left/right)
#R,G,B : 0~255 brightness
tiki.set_led(0, 1, 10, 0, 0)
```

```python
#Turn on/off all 16-bit Red LED.
for i in range(16):
    tiki.set_led(0, i, 0, 10, 0)
time.sleep(2)
for i in range(16):
    tiki.set_led(0, i, 0, 0, 0)
```

**5. Acceleration Sensor**<br>

```python
# Acceleration Sensor Values
# X-axis, Y-axis, Z-axis
tiki.get_imu() 
```

**6. Motor Control - PWM Mode**<br>

```python
# Motor PWM mode setting
# PWM for simple speed control
# rpm : -127 ~ 127
tiki.set_motor_mode(tiki.MOTOR_MODE_PWM)
```

```python
tiki.set_motor_power(tiki.MOTOR_RIGHT, 20)  #-127~127
```

```python
tiki.stop()
```

```python
tiki.set_motor_power(tiki.MOTOR_LEFT, 20)
```

```python
tiki.stop()
```

```python
tiki.forward(20)
```

```python
tiki.forward(20)
time.sleep(2)
tiki.backward(20)
time.sleep(2)
tiki.clockwise(20)
time.sleep(2)
tiki.counter_clockwise(20)
time.sleep(2)
tiki.stop()
```

**7. Motor Control - PID Mode**<br>

```python
# Motor PID mode setting
# PID for precise speed/position control
# rpm : -127 ~ 127
tiki.set_motor_mode(tiki.MOTOR_MODE_PID)
```

```python
tiki.forward(20)
time.sleep(2)
tiki.backward(20)
time.sleep(2)
tiki.clockwise(20)
time.sleep(2)
tiki.counter_clockwise(20)
time.sleep(2)
tiki.stop()
```

**8. Motor Encoder**<br>

```python
left_encoder = tiki.get_encoder(tiki.MOTOR_LEFT)
right_encoder = tiki.get_encoder(tiki.MOTOR_RIGHT)

print(f"Left Encoder: {left_encoder}, Right Encoder: {right_encoder}")
```
