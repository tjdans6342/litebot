```python
#Robot Class package
from tiki.mini import TikiMini
tiki = TikiMini()
```

```python
#Turn on all 16-bit Red LED.
for i in range(16):
    tiki.set_led(0, i, 50, 0, 0)
```

```python
#"X" output
x_idx = [15,0,9,6,10,5,12,3]
for i in x_idx:
    tiki.set_led(0, i, 50,0,0)   # RED
```

```python
#"O" output
o_idx = [15,8,7,0, 12,11,4,3, 14,13, 1,2]
for i in o_idx:
    tiki.set_led(0, i, 0, 50,0)   # GREEN
```

```python
#"#" output
hash_idx = list(set([8,9,10,11, 7,6,5,4, 14,9,6,1, 13,10,5,2]))
for i in hash_idx:
    tiki.set_led(0, i, 0,0,50)# BLUE
```

```python
# 16-bit LED Control Function

from tiki.mini import TikiMini
tiki = TikiMini()

def clear_leds():
    for i in range(16):
        tiki.set_led(0, i, 0, 0, 0)

def draw_shape(shape, color):
    clear_leds()
    r, g, b = color
    if shape == 'X':
        indices = [15,0,9,6,10,5,12,3]
    elif shape == 'O':
        indices = [15,8,7,0, 12,11,4,3, 14,13,1,2]
    elif shape == '#':
        indices = list(set([8,9,10,11, 7,6,5,4, 14,9,6,1, 13,10,5,2]))
    else:
        print("Unknown shape")
        return
    for i in indices:
        tiki.set_led(0, i, r, g, b)
```

```python
draw_shape('X',(50,0,0))
```

```python
draw_shape('O',(0,50,0))
```

```python
draw_shape('#',(0,0,50))
```
