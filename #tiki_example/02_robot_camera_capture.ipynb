```python
import cv2
import ipywidgets as widgets
from IPython.display import display
import os
import time

os.makedirs("capture", exist_ok=True)
file_count = 1

cap = cv2.VideoCapture(
    "nvarguscamerasrc ! video/x-raw(memory:NVMM), width=320, height=240, format=NV12, framerate=60/1 ! "
    "nvvidconv ! video/x-raw, format=BGRx ! videoconvert ! video/x-raw, format=BGR ! appsink",
    cv2.CAP_GSTREAMER
)

if not cap.isOpened():
    print("Camera Open Error")

video_widget = widgets.Image(format='jpeg', layout=widgets.Layout(width='320px', height='240px'))
display(video_widget)

capture_button = widgets.Button(description="Capture")
display(capture_button)

frame = None  # 최신 프레임 저장용

def frame_to_bytes(f):
    _, buf = cv2.imencode('.jpg', f)
    return buf.tobytes()

def capture(b):
    global file_count, frame
    if frame is not None:
        filename = f"capture/test{file_count:02d}.jpg"
        cv2.imwrite(filename, frame)
        print(f"[Captured] {filename}")
        file_count += 1

capture_button.on_click(capture)

try:
    while True:
        ret, f = cap.read()
        if ret:
            f = cv2.flip(f, 0)
            frame = f
            video_widget.value = frame_to_bytes(f)
        time.sleep(0.02)

except KeyboardInterrupt:
    print("Camera stream exit")

finally:
    cap.release()
```
